```python
import pandas as pd
import sqlite3 as sql
import math
from pandas import Series
conn = sql.connect("factbook.db")
facts = pd.read_sql_query("SELECT * FROM facts;", conn)
facts = facts.dropna(axis=0)
facts[:10]
```

```python
c = conn.cursor()
c.execute('SELECT name FROM facts WHERE name!="World" ORDER BY population DESC LIMIT 10;')
print(c.fetchall())
```

```python
def predict_population(row):
    return row["population"]*math.e**(row["population_growth"]/100*35)
new_pop = facts.apply(predict_population, axis=1)
```

```python
facts["new_population"] = new_pop
facts[:10]
```

```python
facts = facts.sort_values(by="new_population", ascending=False)
print(facts["name"][:10])
```

```python
c = conn.cursor()
c.execute('SELECT SUM(area_land) FROM facts WHERE area_land!="";')
area_land_total = c.fetchone()
c.execute('SELECT SUM(area_water) FROM facts WHERE area_water!="";')
area_water_total = c.fetchone()
```

```python
ratio = area_land_total[0]/area_water_total[0]
print(ratio)
```
